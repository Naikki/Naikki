### Calculando o Valor das Prestações

Este algoritmo irá solicitar o nome e o valor total de um produto. Em seguida, ele calculará o valor de cada uma das 5 prestações sem juros e exibirá o nome do produto junto com o valor de cada prestação.

```python
# Solicita o nome do produto ao usuário
nome_do_produto = input('Digite o nome do produto: ')

# Solicita o valor do produto ao usuário como uma string
valor_do_produto_str = input('Digite o valor do produto (use . ou , para decimais, ex: 1500.50 ou 1.500,50): ')

try:
    # Substitui vírgulas por pontos para garantir a conversão correta para float
    valor_do_produto_limpo = valor_do_produto_str.replace('.', '').replace(',', '.')
    valor_do_produto = float(valor_do_produto_limpo)

    # Calcula o valor de cada prestação (dividido em 5 vezes sem juros)
    valor_por_prestacao = valor_do_produto / 5

    # Exibe o resultado
    print(f'\nProduto: {nome_do_produto}')
    print(f'Valor total: R$ {valor_do_produto:.2f}')
    print(f'Valor de cada prestação (5x sem juros): R$ {valor_por_prestacao:.2f}')

except ValueError:
    # Captura o erro se a conversão para float falhar
    print('Erro: Por favor, insira um valor numérico válido para o produto.')
except Exception as e:
    # Captura quaisquer outros erros inesperados
    print(f'Ocorreu um erro inesperado: {e}')
```

### Calculando o Consumo Médio de Combustível

Este algoritmo irá solicitar a distância total percorrida e a quantidade de combustível consumido por um automóvel. Em seguida, ele calculará e exibirá o consumo médio em km/l.

```python
# Solicita a distância total percorrida
distancia_str = input('Digite a distância total percorrida (em km): ')

# Solicita a quantidade de combustível gasta
combustivel_str = input('Digite a quantidade de combustível gasta (em litros): ')

try:
    # Converte as entradas para números de ponto flutuante
    distancia = float(distancia_str.replace(',', '.'))
    combustivel = float(combustivel_str.replace(',', '.'))

    # Verifica se o combustível gasto é maior que zero para evitar divisão por zero
    if combustivel > 0:
        consumo_medio = distancia / combustivel
        print(f'\nDistância percorrida: {distancia:.2f} km')
        print(f'Combustível gasto: {combustivel:.2f} litros')
        print(f'Consumo médio: {consumo_medio:.2f} km/l')
    else:
        print('Erro: A quantidade de combustível gasta deve ser maior que zero.')

except ValueError:
    # Captura o erro se a conversão para float falhar
    print('Erro: Por favor, insira valores numéricos válidos para distância e combustível.')
except Exception as e:
    # Captura quaisquer outros erros inesperados
    print(f'Ocorreu um erro inesperado: {e}')
```

### Calculando Médias de Alunos

Este algoritmo irá solicitar o nome completo de um aluno e suas quatro notas. Com base nessas notas, ele calculará e exibirá a média aritmética anual e as médias aritméticas de cada semestre.

```python
# Solicita o nome do aluno
nome_aluno = input('Digite o nome completo do aluno: ')

# Solicita as quatro notas
nota_1_str = input('Digite a primeira nota: ')
nota_2_str = input('Digite a segunda nota: ')
nota_3_str = input('Digite a terceira nota: ')
nota_4_str = input('Digite a quarta nota: ')

try:
    # Converte as notas para float, substituindo vírgulas por pontos para garantir a conversão
    nota_1 = float(nota_1_str.replace(',', '.'))
    nota_2 = float(nota_2_str.replace(',', '.'))
    nota_3 = float(nota_3_str.replace(',', '.'))
    nota_4 = float(nota_4_str.replace(',', '.'))

    # Calcula a média anual
    media_anual = (nota_1 + nota_2 + nota_3 + nota_4) / 4

    # Calcula as médias semestrais (assumindo 2 notas por semestre)
    media_semestral_1 = (nota_1 + nota_2) / 2
    media_semestral_2 = (nota_3 + nota_4) / 2

    # Exibe os resultados
    print(f'\nAluno: {nome_aluno}')
    print(f'Média Anual: {media_anual:.2f}')
    print(f'Média do 1º Semestre: {media_semestral_1:.2f}')
    print(f'Média do 2º Semestre: {media_semestral_2:.2f}')

except ValueError:
    # Captura o erro se a conversão para float falhar
    print('Erro: Por favor, insira valores numéricos válidos para as notas (use . ou , para decimais).')
except Exception as e:
    # Captura quaisquer outros erros inesperados
    print(f'Ocorreu um erro inesperado: {e}')
```

### Calculando o Preço de Venda com Margem de Lucro

Este algoritmo irá solicitar o preço de custo de um produto e a margem de lucro desejada em porcentagem (%). Em seguida, ele calculará e exibirá o preço de venda final do produto, já incluindo o lucro.

```python
# Solicita o preço de custo do produto
preco_custo_str = input('Digite o preço de custo do produto: ')

# Solicita a margem de lucro em porcentagem
margem_lucro_str = input('Digite a margem de lucro desejada (em porcentagem, ex: 20 para 20%): ')

try:
    # Converte as entradas para números de ponto flutuante
    preco_custo = float(preco_custo_str.replace(',', '.'))
    margem_lucro_percent = float(margem_lucro_str.replace(',', '.'))

    # Verifica se os valores são válidos
    if preco_custo < 0:
        print('Erro: O preço de custo não pode ser negativo.')
    elif margem_lucro_percent < 0:
        print('Erro: A margem de lucro não pode ser negativa.')
    else:
        # Calcula o preço de venda
        # preco_venda = preco_custo + (preco_custo * (margem_lucro_percent / 100))
        # Forma simplificada: preco_venda = preco_custo * (1 + margem_lucro_percent / 100)
        preco_venda = preco_custo * (1 + margem_lucro_percent / 100)

        # Exibe o resultado
        print(f'\nPreço de Custo: R$ {preco_custo:.2f}')
        print(f'Margem de Lucro: {margem_lucro_percent:.2f}%')
        print(f'Preço de Venda (com lucro): R$ {preco_venda:.2f}')

except ValueError:
    # Captura o erro se a conversão para float falhar
    print('Erro: Por favor, insira valores numéricos válidos para o preço de custo e a margem de lucro.')
except Exception as e:
    # Captura quaisquer outros erros inesperados
    print(f'Ocorreu um erro inesperado: {e}')
```

### Calculando o Salário Final de um Vendedor com Comissão

Este algoritmo irá solicitar o nome do vendedor, seu salário base mensal e o valor total de vendas realizadas no mês. Com base nisso, ele calculará o salário final (salário base + 15% de comissão sobre as vendas) e a porcentagem que o salário final representa a mais em relação ao salário base.

```python
# Digitar o nome do Vendedor
nome = input("Digite o nome do vendedor: ")

# Digitar o Salário e o Total de vendas
# .replace('.', '') - Remove pontos que separam milhares (ex: 5.500 → 5500)
salario_base_str = input("Digite o salário base mensal: R$ ").replace('.', '').replace(',', '.')
total_vendas_str = input("Digite o valor total de vendas do mês: R$ ").replace('.', '').replace(',', '.')

# Converte o texto do salário para número decimal se não estiver vazio, caso contrário usa zero.
# Converte o texto das vendas para número decimal se não estiver vazio, caso contrário usa zero.
salario_base = float(salario_base_str) if salario_base_str else 0
total_vendas = float(total_vendas_str) if total_vendas_str else 0

# Calcula a comissão de 15% sobre o total de vendas.
# Soma o salário base com a comissão para obter o salário final.
comissao = total_vendas * 0.15
salario_final = salario_base + comissao

# Calcula a porcentagem que a comissão representa em relação ao salário base. 2a linha.
if salario_base > 0: # Verificar se o salário base é maior que zero para evitar divisão por zero.
    porcentagem_a_mais = (comissao / salario_base) * 100
    aumento_salarial = comissao  # Valor absoluto do aumento
else:
    porcentagem_a_mais = 0
    aumento_salarial = 0
"""Define ambos os valores como zero quando o salário
base for zero ou negativo, evitando cálculo
inválido e garantindo que o programa continue
funcionando sem erros de divisão por zero."""

# Formatação para padrão o brasileiro
def formatar_br(valor):
    return f"{valor:,.2f}".replace(',', 'X').replace('.', ',').replace('X', '.')

print("\n" + "="*50)
print("RESULTADO DETALHADO:")
print("="*50)
print(f"Vendedor: {nome}")
print(f"1. Salário base fixo: R$ {formatar_br(salario_base)}")
print(f"2. Total de vendas: R$ {formatar_br(total_vendas)}")
print(f"3. Comissão (15% das vendas): R$ {formatar_br(comissao)}")
print(f"4. Salário final (1 + 3): R$ {formatar_br(salario_final)}")
print("\nANÁLISE DO AUMENTO:")
print(f"• Valor do aumento (comissão): R$ {formatar_br(aumento_salarial)}")
print(f"• Porcentagem do aumento em relação ao salário base: {porcentagem_a_mais:.2f}%")
print(f"• Ou seja: você recebeu {porcentagem_a_mais:.2f}% a mais este mês")
print("="*50)
```
